```python
import scipy as sp
import scipy.interpolate

def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = sp.interpolate.interp1d(logx, logy, kind=kind, fill_value='extrapolate')
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp
```

```python
def loadSpectrum(mineral):
    x_bins = np.logspace(x_min,x_max,n_bins)

    # Background spectrum for neutrinos
    
    #def dRdx_nu(self,x_bins, components=False, gaussian=False):
    x_width = np.diff(x_bins)
    x = x_bins[:-1] + x_width/2
    #Returns in events/kg/Myr/nm
    #nu_list = ['1987IA']
    nu_list = ['DSNB', 'atm', 'hep', '8B', '15O', '17F', '13N', 'pep','pp','7Be-384','7Be-861', 'K', 'U', '235U', 'Th', '1987IA','CCSN']
    #nu_list = ['DSNB', 'atm', 'hep', '8B', '15O', '17F', '13N', 'pep','pp','7Be-384','7Be-861']
    
    E_list = np.logspace(-3, 3, 5000) # keV
    
    
    dRdx = []
    for j, nu_source in enumerate(nu_list):
        dRdx_temp = np.zeros_like(x)
        for i, nuc in enumerate(mineral.nuclei):
            if (nuc != "H"):
                xtemp = mineral.Etox_nuclei[nuc](E_list)
                dRdx_nuc = (np.vectorize(DMU.dRdE_CEvNS)(E_list, mineral.N_p[i], mineral.N_n[i], flux_name=nu_source)
                                                            *mineral.dEdx_nuclei[nuc](xtemp))
                temp_interp = interp1d(xtemp, dRdx_nuc, fill_value='extrapolate')
                dRdx_temp += mineral.ratio_nuclei[nuc]*temp_interp(x)
        #if(j==11 or j==12 or j==13 or j==14):
            #print(dRdx_temp)
        dRdx.append(dRdx_temp*1e6*365+1e-20)
        #dRdx.append(dRdx_temp*365+1e-20)
        
    solar = np.zeros(n_bins-1)
    dRdx.append(solar)
        
    for i in range(n_bins-1):
        for j in range(9):
            dRdx[17][i] += dRdx[j+2][i]
            
    geo = np.zeros(n_bins-1)
    dRdx.append(geo)
    

        
    for i in range(n_bins-1):
        for j in range(4):
            dRdx[18][i] += dRdx[j+11][i]
            #print(dRdx[j+11])
    return dRdx
```

```python
def Integrate(dRdx, sigma, mass, t, omega = 1):
    
    x_bins_all = np.logspace(x_min, x_max, n_bins)
    x_width_all = np.diff(x_bins_all)
    x_mid_all = x_bins_all[:-1] + x_width_all/2.
    
    # bins biased
    x0 = sigma/2.0
    x_bins = np.logspace(x_min, x_max, n_bins)
    if(sigma != 0):
        x_bins = np.logspace(np.log10(x0), x_max, n_bins)
    x_width = np.diff(x_bins)
    x_mid= x_bins[:-1] + x_width/2.
    
    dRdx_interp = interp1d(x_mid_all, dRdx, bounds_error=False, fill_value=0.0)
    
    N_events_ind = []
    N = 2
    for i in range(n_bins - 1):
        x1 = x_bins[i]-5*sigma
        x2 = x_bins[i+1]+5*sigma
        m = (x1+x2)/2.0
        
        if(sigma != 0):
            integrand = lambda x: dRdx_interp(x)*paleopy.Max(x, x_bins[i], x_bins[i+1], sigma, N)
        else:
            integrand = lambda x: dRdx_interp(x)
        
        A = quad(integrand, x1, x2, epsrel=1e-4)[0]*t*mass*omega
        
        N_events_ind.append(A)
        
    z = np.linspace(h_min, h_max, h_bins)
    y = []
    
    for i in range(h_bins-int(sigma/2*10)):
        for j in range(n_bins - 1):
            if(z[i+int(sigma/2*10)]<=x_bins[j+1]):
                y.append(N_events_ind[j])
                break
                
    print("Finito!")
   
    return y
```

```python
def IntegrateTh(sigma):
    
    if(sigma != 0):
    
        T_exp = 1e9
        tau_Th = 1.405e+10/np.log(2)
        x0 = sigma/2.0
        x_bins = np.logspace(np.log10(x0), x_max, n_bins)
    
        N = 2
        Nevents_Th = np.zeros(n_bins-1)
        x_Th = mineral.xT_Thorium()
        for i in range(n_bins-1):
            Nevents_Th[i] = paleopy.Max((x_bins[i]+x_bins[i+1])/2, x_Th-1, x_Th+1, sigma, N)
            #Nevents_Th[i] = paleopy.window(x_Th, x_bins[i], x_bins[i+1], sigma)*tau_Th*(exp(t*1e6/tau_Th)-1)/1e6*massa
        
        Nevents_Th = mineral.norm_Thorium(T=T_exp)*Nevents_Th
    
        z = np.linspace(h_min, h_max, h_bins)
        y = []
    
        for i in range(h_bins-int(sigma/2*10)):
            for j in range(n_bins-1):
                if(z[i+int(sigma/2*10)] <= x_bins[j+1]):
                    y.append(Nevents_Th[j])
                    break
                
        print("Finito!")
   
        return y, 0

    else:
        
        x_bins = np.logspace(x_min, x_max, n_bins)
        x_Th = mineral.xT_Thorium()
        x = np.linspace(h_min,h_max,h_bins)
        
        y = np.zeros(h_bins)
        index = 0
        
        for i in range(n_bins-1):
            for j in range(h_bins):
                if(x[j] < x_bins[i+1] and x[j] >= x_bins[i]):
                    if(x_Th < x_bins[i+1] and x_Th >= x_bins[i]):
                        index = i
                        y[j] = mineral.norm_Thorium(T=1e7)
        
        return y, index       
```

```python
def Integral(dRdx, x_bins):
    x_width_all = np.diff(x_bins)
    x_mid_all = x_bins[:-1] + x_width_all/2.
    
    dRdx_interp = interp1d(x_mid_all, dRdx, bounds_error=False, fill_value=0.0)
    A = []
    for i in range(len(x_bins)-1):
        A.append(quad(dRdx_interp,x_bins[i],x_bins[i+1])[0])
    return A
```

```python
def IntegralTh(x_bins):
    x_Th = mineral.xT_Thorium()
    x = np.linspace(h_min,h_max,h_bins)
    
    A = []
    
    for i in range(len(x_bins)-1):
        if(x_Th>=x_bins[i] and x_bins[i+1]>x_Th):
            A.append(mineral.norm_Thorium(T=1e7))
        else: A.append(0)
        
    return A
```

```python
def IntegralSSF(name,x_bins):
    x,y  = np.loadtxt("Data/"+name+"/"+name+"_fission.dat", usecols = (0,1), unpack='true')
    Ufrac = 5*1e4
    
    fiss = interp1d(x, Ufrac*y, fill_value='extrapolate')
    
    A = []
    for i in range(len(x_bins)-1):
        A.append(quad(fiss,x_bins[i],x_bins[i+1])[0])
    return A
```

```python
import scipy as sp
import scipy.interpolate

def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = sp.interpolate.interp1d(logx, logy, kind=kind, fill_value='extrapolate')
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp
```
